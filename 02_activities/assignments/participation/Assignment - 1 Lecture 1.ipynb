```python
#Making a basic figure with matplotlib
#Loading Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import PIL as pil
import requests as req
```

```python
#Making sample data to plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
#Make a basic scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#Modifying the graph
fig,ax=plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
#Modifying the graph to be horizontal
fig,ax=plt.subplots(figsize=(5, 3))
ax.barh(x,y) #This will make it horizontal
```

```python
#Making a line plot
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y) #Just writing ax.plot will make a line plot
```

```python
#Activity: Try and make a histogram
fig, ax=plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
#Adding axis labels and titles
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y) #This will make a line plot
ax.set_title('Total growth over time') #This is the title which is a parameter in the brackets
ax.set_ylabel('Total Growth') #This is the y axis label
ax.set_xlabel('Years since start') #This is the x axis label
fig.tight_layout() #This will make sure the labels are not cut off
```

```python
#Modifying labels and titles
#need to use font dictionaries
font1 = {'family':'serif','color':'blue','size':20} #numbers are not in quotations
font2 = {'family':'monospace','color':'green','size':14}
```

```python
#to add the edits for labels and titles, we need to add the font dictionaries
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict=font1) 
ax.set_ylabel('Total Growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2) 
```

```python
#Moving labels and titles
#we can move labels and titles by using the loc argument
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time',fontdict=font1, loc='left')
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('Years since start',fontdict=font2)  
fig.tight_layout()
```

```python
#Modifying data points
#Lets remake the basic scatterplot but add colour and marker arguments
#Needs to be american spelling
fig, ax=plt.subplots(figsize=(5, 3))
ax.scatter(x,y,marker='*', color='indigo')
```

#There are other markers that can be used as seen here
https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

```python
#Line width and line style modification
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',color='indigo',
        linestyle='--', linewidth=2) #This will make the line dashed and thicker
fig.show()
```

```python
#Modifying data points - colour
#We were previously naming the colours. However, we can also use hexcodes
fig, ax=plt.subplots(figsize=(5, 3))
#colour is the new addition
ax.plot(x,y,marker='*',color='#7425b9',
        linestyle='--',linewidth=2)
fig.show()
```

```python
#modifying marker size, edge colour and face colour
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
        markersize=12, #This makes the marker larger
        color='#7425b9',
        linestyle='--',linewidth=2,
        markeredgecolor='#fa9359', #This is the edge colour
        markerfacecolor='#000000') #This is the face colour
```

```python
#Adding grid lines using .grid()
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
        markersize=12, #This makes the marker larger
        color='#7425b9',
        linestyle='--',linewidth=2,
        markeredgecolor='#fa9359', #This is the edge colour
        markerfacecolor='#000000') #This is the face colour
ax.grid()
```

```python
#modify the grid lines
fig, ax=plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
        markersize=12, #This makes the marker larger
        color='#7425b9',
        linestyle='--',linewidth=2,
        markeredgecolor='#fa9359', #This is the edge colour
        markerfacecolor='#000000') #This is the face colour
ax.grid(axis='y',
        color='blue',
        linewidth=2,linestyle='--')
```
