Making Visualisation with Seaborn

```python
#First install the libraries
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import PIL as pil
import requests
```

```python
#loading data 
tips = sns.load_dataset("tips")
print(tips)
```

```python
#Building a basic plot
#Line plot of tip amount vs total bill
sns.lineplot(data=tips, #choose dataset
             x='total_bill', #define x variable
             y='tip') #define y variable
```

```python
#Modifying figures using styles
sns.set_style('whitegrid') #white background with x and y grid
sns.lineplot(data=tips, #choose dataset
             x='total_bill', #define x variable
             y='tip') #define y variable
```

```python
#Modifying figures using styles
sns.set_style('darkgrid') #white background with x and y grid
sns.lineplot(data=tips, #choose dataset
             x='total_bill', #define x variable
             y='tip') #define y variable
```

```python
#Adding titles and labels
tipgraph = sns.lineplot(data=tips, #choose dataset
                x='total_bill', #define x variable
                y='tip') #define y variable
tipgraph.set(title= 'Tips vs Total Bill', #title of graph
             xlabel = 'Total Bill ($)', #x axis label
             ylabel = 'Tip ($)') #y axis label
```

```python
#Modifying aesthetics
fig = plt.subplots(figsize=(10,3))
tipgraph = sns.lineplot(data=tips, #choose dataset
                x='total_bill', #define x variable
                y='tip') #define y variable
tipgraph.set(title= 'Tips vs Total Bill', #title of graph
             xlabel = 'Total Bill ($)', #x axis label
             ylabel = 'Tip ($)') #y axis label
```

```python
#Changing colour, marker style and line style
fig = plt.subplots(figsize=(10,3)) 

tipgraph = sns.lineplot(data=tips, #choose dataset
                x='total_bill', #define x variable
                y='tip',
                   color = 'hotpink',#background colour
                   linestyle ='dashed',
                   linewidth = 3,
                   marker = 'o',
                   markerfacecolor = 'indigo') #define y variable
tipgraph.set(title= 'Tips vs Total Bill', #title of graph
             xlabel = 'Total Bill ($)', #x axis label
             ylabel = 'Tip ($)') #y axis label
```

```python
#Adding multiple variabbles
tipgraph = sns.scatterplot(data=tips,
                           x='total_bill',
                           y='tip',
                           style = 'time',
                           hue = 'day',
                           palette =['purple','hotpink','deepskyblue','yellowgreen'])
tipgraph.set(title= 'Tips vs Total Bill', #title of graph
                xlabel = 'Total Bill ($)', #x axis label
                ylabel = 'Tip ($)') #y axis label
```

```python
#Pairplot
sns.pairplot(data=tips, hue='day')
```

```python
#Relplot - explore specific levels within variables
daysplot = sns.relplot(data=tips, 
                       x='total_bill',
                       y='tip',
                       hue='sex',
                       col='day',
                       kind='scatter',
                       col_wrap=2) #define number of columns
```

```python
#Setting up 
import plotly.graph_objects as go #go is graph objects
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted


x1 = np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1 = np.array([110,180,240,99,220])
```

```python
#Making our plot
#PLotly syntax is different from matplotlib and seaborn
graph = go.Figure()
#Then choose our plot type (bar) and our data
graph.add_trace(go.Bar(x=x1,y=y1))
```

```python
#Adding titles
graph = go.Figure()
graph.add_trace(go.Bar(x=x1,y=y1))
graph.update_layout(title_text='Pirate Scores',
                    xaxis_title_text='Pirate',
                    yaxis_title_text='Score')
graph.show()
```

```python
#Customising plotly graphs
#Change the bar chart into scatter plot - go.scatter
#we can specify the mode as markers
#Change the look of the markers such as colour, opacity, outer outline
#We are going to update the tiotle and axis title layout.
#We are going to specify the size using width and height parameters

graph = go.Figure()
graph.add_trace(go.Scatter(x=x1,y=y1,
                mode='markers',
                marker=dict(size=15,
                            color = 'hotpink',
                            opacity = 1,
                            line =dict(width=5,
                                       color='purple')
                                       ))) #the dictionary is not similar to normal dictionaries
#adding title first 
graph.update_layout(
    title='Interactive Pirate Plot',
    xaxis_title='Pirate',
    yaxis_title='Score',
    width=500, #width of the graph
    height=500) #height of the graph
```

Wordclouds and Venn Diagrams

```python
#remember to load the wordcloud package
from wordcloud import WordCloud
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv',
                 on_bad_lines='skip')#loads the dataset. The on_bad_lines is to skip the bad lines in the dataset
#if there are any rows of data that has more columns than expected, it will skip that data
#we can also say 'error' so it writes it as an erro message
df
```

```python
text = " ".join(each for each in df.quote) #make sure that there is a space between the quotatiosn
wordcloud = WordCloud(background_color='white',
                      colormap='inferno').generate(text) #generate the wordcloud using the text 
fig, ax=plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,interpolation='bilinear')
ax.axis('off') #turn off the axis
plt.show()
```

Making Venn Diagrams

```python
#First import the package
#from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
#Then define our sets
A = set(['apple','banana','watermelon'])
B = set(['pumpkin','blueberry','banana','keylime'])
```

```python
diagram = venn2_unweighted([A,B], #the two sets
                           set_labels=('Fruits','Pies'),
                           set_colors=('blue','red'),
                           alpha=0.5) #the alpha is the transparency of the colours
plt.show()
```

```python
#Modifying Venn Diagrams
diagram = venn2_unweighted([A,B], #the two sets
                           set_labels=('Fruits','Pies'),
                           set_colors=('blue','red'),
                           alpha=0.5) #the alpha is the transparency of the colours
diagram.get_label_by_id('10').set_text('\n'.join(A - B)) #the 10 is the first set, 
diagram.get_label_by_id('11').set_text('\n'.join(A & B)) #the 11 is the intersection of both sets
diagram.get_label_by_id('01').set_text('\n'.join(B - A)) #the 01 is the second set

plt.show()
```

```python
#REmember we need to have the \n otherwise the text will be on the same line
#Modifying Venn Diagrams
diagram = venn2_unweighted([A,B], #the two sets
                           set_labels=('Fruits','Pies'),
                           set_colors=('blue','red'),
                           alpha=0.5) #the alpha is the transparency of the colours
diagram.get_label_by_id('10').set_text('n'.join(A - B)) #the 10 is the first set, 
diagram.get_label_by_id('11').set_text('n'.join(A & B)) #the 11 is the intersection of both sets
diagram.get_label_by_id('01').set_text('n'.join(B - A)) #the 01 is the second set

plt.show()
```
